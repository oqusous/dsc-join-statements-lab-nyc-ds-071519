## Display the names of all the employees in Boston.
Hint: join the employees and offices tables.

```python
#Your code here
c.execute("""SELECT firstname, lastname, city FROM employees e
             LEFT JOIN offices o
             ON e.officeCode = o.officeCode
             WHERE city like '%Boston%';""")
df = pd.DataFrame(c.fetchall())
df.columns = [x[0] for x in c.description]
df
```

## Do any offices have no employees?
Hint: Combine the employees and offices tables and use a group by.

```python
#Your code here
c.execute("""SELECT count(firstname), city FROM employees e
             LEFT JOIN offices o
             ON e.officeCode = o.officeCode
             GROUP BY city;""")
df = pd.DataFrame(c.fetchall())
df.columns = [x[0] for x in c.description]
df
```

## Level Up: Display the Number of Products each employee has sold

```python
# Your code here
c.execute("""SELECT count(productName) as Number_of_products, e.employeeNumber, e.firstName, e.lastName FROM products p
             LEFT JOIN orderdetails od
             ON od.productCode = p.productCode
             LEFT JOIN orders o
             ON od.orderNumber = o.orderNumber
             LEFT JOIN customers c
             ON c.customerNumber = o.customerNumber
             LEFT JOIN employees e
             ON e.employeeNumber = c.salesRepEmployeeNumber
             GROUP BY e.employeeNumber
             order by Number_of_products DESC;""")
df = pd.DataFrame(c.fetchall())
df.columns = [x[0] for x in c.description]
df
```
